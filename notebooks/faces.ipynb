```python
import utils.cv_utils as cvu
import cv2 as cv
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("../opencv-course/Resources/Photos/lady.jpg")
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
cvu.cv_imshow("Gray Person", gray)
```

```python
haar_cascade = cv.CascadeClassifier(f'{cv.data.haarcascades}haarcascade_frontalface_default.xml')
```

```python
faces_rectangle = haar_cascade.detectMultiScale(image=gray, scaleFactor=1.1, minNeighbors=3)
```

```python
print(f"{len(faces_rectangle)} found in this image")
```

```python
for (x, y, w, h) in faces_rectangle:
    cv.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), thickness=2)
    
```

```python
cvu.cv_imshow("Gray Person", img)
```

```python
img = cv.imread("../opencv-course/Resources/Photos/group 1.jpg")
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
faces_rectangle = haar_cascade.detectMultiScale(image=gray, scaleFactor=1.1, minNeighbors=3)
print(f"{len(faces_rectangle)} found in this image")
for (x, y, w, h) in faces_rectangle:
    cv.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), thickness=2)
cvu.cv_imshow("Gray Person", img)   
```
